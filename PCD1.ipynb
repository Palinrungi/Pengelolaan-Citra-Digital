```python
# Original image
im_original = im.copy()

# Bright image
im_bright = rgb2lab(im)
im_bright[...,0] = im_bright[...,0] + 50  # Tingkatkan kecerahan
im_bright = lab2rgb(im_bright)

# Darker image
im_darker = rgb2lab(im)
im_darker[...,0] = im_darker[...,0] - 50  # Kurangi kecerahan
im_darker = lab2rgb(im_darker)
```

```python
# Tampilkan gambar
plt.figure(figsize=(20,10))
plt.subplot(131), plt.imshow(im_original), plt.axis('off'), plt.title('Original Image', size=20)
plt.subplot(132), plt.imshow(im_bright), plt.axis('off'), plt.title('Brighter Image', size=20)
plt.subplot(133), plt.imshow(im_darker), plt.axis('off'), plt.title('Darker Image', size=20)
plt.show()
```

```python
# Gambar asli
im_original = im.copy()

# Gambar terbalik dalam ruang warna LAB
im_lab_inverted = rgb2lab(im)
im_lab_inverted[..., 0] = np.max(im_lab_inverted[..., 0]) - im_lab_inverted[..., 0]

# Gambar terbalik dalam ruang warna RGB
im_rgb_inverted = lab2rgb(im_lab_inverted)

# Tampilkan gambar-gambar
plt.figure(figsize=(20, 10))

plt.subplot(131), plt.imshow(im_original), plt.axis('off'), plt.title('Gambar Asli', size=20)
plt.subplot(132), plt.imshow(im_lab_inverted), plt.axis('off'), plt.title('Gambar Terbalik (LAB)', size=20)
plt.subplot(133), plt.imshow(im_rgb_inverted), plt.axis('off'), plt.title('Gambar Terbalik (RGB)', size=20)

plt.show()
```
